## Absolute and relative paths

---

Python scripts are executed in a shell. Said shells have an active directory (in our case, the folder NEW_3RD_SEMESTER), which is nothing more than the folder we are working on. If we want to open a file through shell commands (or python scripts that execute shell commands) we need to specify the **path** of the file. To do so we have two options: (1) Use an **absolute path** or (2) use a **realtive path**.

### What is an absolute path?

An **absolute path** is nothing more than the full tree-like path of the file, starting from the storage unit that contains it, an example of such a path is `C:/Users/ossia/Documents/SEMESTRES_FACULTAD/New_3rd_semester/INPUTS/INE_ENTIDAD_2020.csv`. 

### What is a relative path?

A  **relative path** relies on the the active/working directory of the shell through which we execute the code. If the file we want is inside the active folder, then only the name of the file along with its extension is necessary. If the file is not, then we need to give its relative path, i.e. its path starting from the point of view of the working directory. To move thorough folders we can use shell commands such as `../` or `/INPUTS/`. The former specifies that the file in the parent directory of our current folder, the latter that the file is in a child directory (a nested folder inside our active one). After each of this commands, we place the name and the extension of the file in question; an example would be `../INPUTS/INE_ENTIDAD_2020.CSV`.

### Which is best?

Which is best will depend on the context, although most of the time using relative paths will prove to be the most optimal. To ensure such optimality, specific data storage frameworks are developed. We, for example, defined 3 folders in the parent directory for the class:

- INPUTS
- OUTPUTS
- SCRIPTS

This way we mantain our files organized and we have an easy way of identifying where a file can be.

### Pandas Series

---

##### **What is a Series?** 

A Series is a one-dimenisonal array-like object containing a **sequence of values** of the same type **and an associated array of data labels**, called its **index**. We can think of Serias as fixed length, ordered dictionaries (for they are essentially mapping the index values to the data values).

### Key IDs

---

Cuandro trabajamos con bases de datos (especialmente bases de datos relacionales) cada fila contiene una serie de valores para un mismo objeto. La identidad de dicho objeto siempre esta relacionada con un ID llave, el cual es unico. Este ID suele ser un tipo de dato mas simple, eficientizando las llamadas a los datos, y facilitando la relacion entre distintas tablas de datos dentro de la base de datos.

Dicho ID llave suele respetar el formato que se tiene para los demas datos de la tabla (mayus o minus dependiendo de los demas datos). El estandar es que sea tipo caracter.

Para esta base datos, ENT es nuestro key ID.

Lo primero que hacemos es verificar que tipo de dato tiene la Series asociada a la columna de "ENT" que en nuestro Key ID.

```python
# Verificamos el tipo de dato de df["ENT"]
print(type(df["ENT"][0]))
```

## What is a dashboard? - Finish

---

A dashboard is a way of displaying various types of visual data in one place. Usually, a dashboard is intended to convey different, but related information in an easy to read way. The most common use case for dashboards are includes  key performance indicators (KPIs) and other important business metrics that decision makers use in theyr day to day activities.

The concept of dashboards may be confused the concept of data visualization. Dashboards are a neat way of visualizing data, they focus on being appealing to the eye, being interactive and connecting several data bases. That is, dashboards are one aspect of the more ample domain of data visualization.

## What is an LLM? - Finish

The acronym 'LMM' stand for Large Language Model. The concept become increasingly popular (with the general population) with the realease of chatgpt by OpenIA (leader in the AI industry). This type of models are a category of deep learning models trained on immense amounts of data, making them capable of understanding and generating natural language and other types of content to perform a wide range of tasks. They are built on a type of neural network (a complicated computational model that mimics the way neurons work in the human brain).

This type of models do not really understand the topics they are able to discuss with enough training. Still, they are extremelly powerfull statistical prediction machines whose whole job is to predict the next word in a sequence.
