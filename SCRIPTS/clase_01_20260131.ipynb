### What are ipython notebooks? 

Ipython notebooks are interactive environments that merge text in markdown and runnable code. Ipython notebooks allow block-like coding, add static impressions and have markdown capabilities.

### What is the advantage of using ipython notebooks?


It enables a notebook-like code development focusing on shareability and readability. Throughout their use, documenting code and performing analysis made through the execution of said code is made easier and more readable by having it all in one single file.

## Shell commands in ipython code blocks

---

Even if a python kernel is selected for a specific ipython notebook, it is possible to run scripts of other programming languages inside code blocks. To know more about this remit to [ipython_magic](https://python4data.science/en/latest/workspace/ipython/magics.html). For now, we'll just explore how this can be used to run powershell or bash commands or scripts.

### How can we run shell commands/scripts?

By adding the symbol '**%**' before a one line command in a code block, we signal ipython to execute **line magics** (a type of magic command), after which we can write a single line of code (in this case, a shell command) and after running the command block, the command will be executed.

Alternatively, to used shell scripts, we can use **cell magics** which is another type of magic command that sets the interpreter of a cellblock, allowing to run entire scripts. To so do, at the top of the code block we need to have the sybol '**%%**' immediatelly followed (without spaces) by the name of the language/interpreter; e.g. `%%bash`, `%%powershell`.

### Ejercicio 3:

Generar una lista diferente por cada tipo de valor

Solution:

```python
# Ejercicio 3: Generar una lista diferente por cada tipo de valor

lista_ej2 = [False,"p",10,11,True,3.1415,"x","1","13",4,48,False,"h",91,2,1.3]

diccionario_types = {}

for elemento in lista_ej2:
    tipo = type(elemento)
    if tipo not in diccionario_types:
        diccionario_types[tipo] = []
    diccionario_types[tipo].append(elemento)

print(diccionario_types)
```

### Ejercicio tarea

De la lista anterior (ejercicio 3), si es entero obtener la raíz cuadrada, si es flotante dividir entre 2, si es booleano obtener el inverso si es string convertirlo a mayúsculas y concatenarle "_STRING" al final. El resultado final debe ser la misma lista con las transformaciones de cada tipo de dato.

Solucion:

```python
lista_ej2 = [False,"p",10,11,True,3.1415,"x","1","13",4,48,False,"h",91,2,1.3]
lista_nueva = []
for elemento in lista_ej2:
    if type(elemento) is bool:
        lista_nueva.append(not elemento)
    elif type(elemento) is int:
        lista_nueva.append(elemento**(1/2))
    elif type(elemento) is float:
        lista_nueva.append(elemento/2)
    elif type(elemento) is str:
        lista_nueva.append(elemento.upper() + "_STRING")
print(lista_nueva)
```

## Git and GitHub - polish details

---

- What is GitHub?
	- GitHub is a cloud-based platform that runs on [[Git]]
- What is GitHub used for?
	- Showcase code
	- Collaborate on code
	- Review code
	- Track and manage code
- How do Git and GitHub work together?
	- When storing files in GitHub, for every change made (or 'commits') to the files on GitHub, Git will automatically track and manage the changes.
	- There are plenty of Git related actions that one can do on the webpage of GitHub, however, most people work on their files locally, and then sync these changes with the central 'remote' repository on GitHub. 
	- This is embodied mainly by push and pulls.
		- Git is the one that intelligently merges this flow of changes, while GitHub provides non-git capabilities such as 'pull requests'.
- On GitHub, saved changes are called commits
	- Each commit has an associated commit message and a Git hash (also called 'commit hash' or just 'hash')
		- The commit message is a description explaining why a particular change was made.
			- These capture the history of changes, easing collaboration so other collaborators can then understand what has been done and why.
		- The commit hash, is a unique identified of a commit.
- Branching
	- What is branching?
		- Just doing commits on main would create a linear history of changes of the files in a repo. Branching allows us to have different versions of a repository at the same time. It provides a tree-like structure for the history of changes of the files in the repo.
	- When are branches used?
		- If we want to make changes but do not want to change the source code, we use branches. Branches are useful for testing new features, as the work done of them will not show up nor affect the main branch (the source code), until we merge said branch into main.
	- What is a branch?
		- A branch is a copy, or snapshot, of another branch as it was in that point in time. The changes made in a branch are not visible in any other, for essentially we are working on a copy.
		- If some changes are made in other branches, we can pull those changes to the desired branch, updating the contents.
	- The main branch
		- By default, any repo has one branch called main. Said branch is considered to be the definitive branch of the repository.
	- The journey a branch takes from conception until it is merged into main
		- The usual journey that a branch takes from conception until it is merged into main or other branch, is the following:
		- 1st. Commit changes
		- 2nd. Submit a Pull request
		- 3rd. Discuss proposed changes
- Pull requests
	- What are pull requests?
		- Pull requests are the hearth of collaboration on GitHub. Once some changes have been done in a specific branch of a repo, we can propose those changes to merged with into another branch, and in the process we request others (including oneself) to review those changes, and pull  in the contribution (the changes) and merge them into said other branch.
	- Some specifics of pull requests
		- On github, Pull requests show diffs, or differences, of the content from both branches. The changes, additions and subtractions are shown in different colors.
		- Pull requests can be submitted before the code (or changes) is finished. In this scenarios, after some discussion of the proposed changes, some further changes (commits) can be made in the branch that we want to merge, and eventually merge it into the main or other branch
		- There is a lot more to cover regarding pull request reviews
- Difference between downloading, cloning and forking
	- Download
		- To download is to save the files of a specific snapshot (or branch) at a given time to your local computer.
		- This is done when we are interesting in doing things with the files, but do not want to apply version control in them.
	- Cloning
		- To clone a repo is to save a copy of it to your local computer.
		- This is done if we want to work on a full copy of the repository in our local computer, using git for version control. Usually, the changes made locally are then synched with the GitHub hosted repo.
	- Fork
		- A fork is a server-side 'clone' of a repository. It creates a new, distinct copy of a repo. This new copy will be liked to the account of whomever forked the repo, so they will have the available permissions to make any further changes.
		- Usually this is done if we want to use the original repository's data as a basis for our own project, or if we want to collaborate on said project. In the former case, commits would be done in the forked repo, and then a pull request would be sent to the 'upstream' repo (i.e. the original proyect).
		- It is also possible and common to clone a fork of a repo
- [[Git workflows]]
- GitHub issues
	- GitHub issues are native to GitHub
		- Is a layer on the GitHub website, the issues will not come when fetching the changes on the 'remote'.
	- Issues are a place to track bug reports, new features and ideas, and anything else you need to write down or discuss with your team.
		- Usually known as GitHub's bug tracker.
	- Only the person who submitted the issue or the owner of the repo can close an issue.
	- All issues have an id number.
		- In the commit messages, we can make a reference to issues by mentioning 'as per \#issue_number', and in the issues section this will be visible.
		- By placing 'fixes \#issue_number', the issue with the id number mentioned is closed.
	- In an issue we can reference commits using their hash code (or using their references?). This can be done by placing the hash of the commit in the message of the issue.
	- Referencing issue numbers and commit hash codes in the commit messages or in the issues descriptions.
	- There is a lot more to learn regarding GitHub issues.
